```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Orthogonal Vectors

Dua vector dikatakan orthogonal
jika dot product bernilai nol.

```python
a = np.array([1,0])
b = np.array([0,1])

dot = np.dot(a,b)

print("Vector a:")
print(a)

print("\nVector b:")
print(b)

print("\nDot Product:")
print(dot)
```

Karena hasil dot product = 0,
maka kedua vector orthogonal.

# 2. Visualization of Orthogonal Vectors

```python
a = np.array([1,0])
b = np.array([0,1])

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    a[0],a[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    b[0],b[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(-1,2)
plt.ylim(-1,2)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Orthogonal Vectors")

plt.show()
```

Grafik menunjukkan dua vector
yang saling tegak lurus.

# 3. Orthogonal Matrix

Orthogonal matrix adalah matrix
yang memenuhi:

QᵀQ = I

```python
Q = np.array([
    [0,-1],
    [1,0]
])

result = Q.T @ Q

print("Matrix Q:")
print(Q)

print("\nQᵀQ:")
print(result)
```

Karena hasilnya identity matrix,
maka Q adalah orthogonal matrix.

# 4. Orthogonal Matrix Inverse

Pada orthogonal matrix:

Q⁻¹ = Qᵀ

```python
Q = np.array([
    [0,-1],
    [1,0]
])

Q_inv = np.linalg.inv(Q)

print("Transpose:")
print(Q.T)

print("\nInverse:")
print(Q_inv)
```

Transpose orthogonal matrix
sama dengan inverse-nya.

# 5. Rotation Matrix as Orthogonal Matrix

Rotation matrix merupakan contoh
orthogonal matrix.

```python
theta = np.pi / 4

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

print("Rotation Matrix:")
print(R)

print("\nRᵀR:")
print(R.T @ R)
```

Rotation matrix mempertahankan
panjang vector setelah transformasi.

# 6. Visualization of Rotation

```python
theta = np.pi / 4

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

v = np.array([1,0])

rotated = R @ v

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    rotated[0],rotated[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(-1,2)
plt.ylim(-1,2)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Orthogonal Rotation")

plt.show()
```

Vector diputar tanpa mengubah
panjang vector.

# 7. Orthonormal Basis

Orthonormal basis terdiri dari:
- vector orthogonal,
- panjang vector = 1.

```python
e1 = np.array([1,0])
e2 = np.array([0,1])

print("Norm e1:")
print(np.linalg.norm(e1))

print("\nNorm e2:")
print(np.linalg.norm(e2))
```

Kedua vector memiliki:
- norm = 1
- saling orthogonal

sehingga membentuk orthonormal basis.

# 8. Projection onto Orthogonal Vectors

```python
a = np.array([3,2])
b = np.array([1,0])

projection = (
    np.dot(a,b) / np.dot(b,b)
) * b

print("Projection:")
print(projection)
```

Projection menghasilkan komponen vector
yang searah dengan vector basis.

# 9. Visualization of Projection

```python
a = np.array([3,2])
b = np.array([1,0])

proj = (
    np.dot(a,b) / np.dot(b,b)
) * b

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    a[0],a[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    proj[0],proj[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,4)
plt.ylim(0,4)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Vector Projection")

plt.show()
```

Projection membantu memahami
komponen vector dalam ruang tertentu.

# 10. Orthogonality in Machine Learning

Orthogonality digunakan dalam:

- PCA
- SVD
- Signal Processing
- Computer Vision
- Deep Learning

```python
X = np.random.randn(5,5)

Q, R = np.linalg.qr(X)

print("Orthogonal Matrix Q:")
print(Q)

print("\nQᵀQ:")
print(Q.T @ Q)
```

QR decomposition menghasilkan
orthogonal matrix Q.

# 11. Heatmap of Orthogonal Matrix

```python
plt.figure(figsize=(5,4))

plt.imshow(Q)

plt.colorbar()

plt.title("Orthogonal Matrix Heatmap")

plt.show()
```

Heatmap membantu visualisasi
struktur orthogonal matrix.
