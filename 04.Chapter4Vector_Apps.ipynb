```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Vectors as Data

Dalam data science,
data sering direpresentasikan sebagai vector.

Contoh:
- tinggi badan
- berat badan
- nilai mahasiswa
- harga saham

Semua dapat direpresentasikan
dalam bentuk vector.

```python
height = np.array([150,155,160,165,170])
weight = np.array([50,55,60,65,70])

print("Height:")
print(height)

print("\nWeight:")
print(weight)
```

Setiap elemen vector merepresentasikan
data dari individu yang sama.

# 2. Visualizing Data Vectors

```python
height = np.array([150,155,160,165,170])
weight = np.array([50,55,60,65,70])

plt.figure(figsize=(6,5))

plt.scatter(height,weight)

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")

plt.grid()

plt.show()
```

Grafik scatter plot menunjukkan hubungan
antara tinggi badan dan berat badan.

# 3. Dot Product in Data Science

Dot product dapat digunakan
untuk mengukur hubungan antar data.

```python
height = np.array([150,155,160,165,170])
weight = np.array([50,55,60,65,70])

dot_product = np.dot(height,weight)

print("Dot Product:")
print(dot_product)
```

Semakin besar hubungan antar vector,
semakin besar nilai dot product.

# 4. Cosine Similarity

Cosine similarity digunakan
untuk mengukur kemiripan arah antar vector.

Konsep ini banyak digunakan pada:
- recommendation system
- search engine
- natural language processing

```python
v1 = np.array([1,2,3,4])
v2 = np.array([2,4,6,8])

cosine_similarity = np.dot(v1,v2) / (
    np.linalg.norm(v1) * np.linalg.norm(v2)
)

print("Cosine Similarity:")
print(cosine_similarity)
```

Nilai cosine similarity mendekati 1
menunjukkan kedua vector sangat mirip.

# 5. Pearson Correlation

Pearson correlation digunakan
untuk mengukur hubungan linear antar data.

```python
height = np.array([150,155,160,165,170])
weight = np.array([50,55,60,65,70])

correlation = np.corrcoef(height,weight)

print(correlation)
```

Nilai correlation mendekati 1
menunjukkan hubungan linear positif kuat.

# 6. Visualizing Correlation

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

plt.figure(figsize=(6,5))

plt.plot(x,y,'o-')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Positive Correlation")

plt.grid()

plt.show()
```

Grafik menunjukkan hubungan linear positif.
Semakin besar x,
semakin besar y.

# 7. Orthogonal Vectors in Data Science

Orthogonal vector menunjukkan
dua data tidak memiliki hubungan linear.

```python
a = np.array([1,0])
b = np.array([0,1])

dot = np.dot(a,b)

print("Dot Product:")
print(dot)
```

Karena dot product = 0,
maka kedua vector orthogonal.

# 8. Projection of Data

Projection digunakan untuk:
- feature extraction
- dimensionality reduction
- machine learning

```python
a = np.array([4,2])
b = np.array([1,1])

projection = (
    np.dot(a,b) / np.dot(b,b)
) * b

print("Projection:")
print(projection)
```

Projection menghasilkan bagian vector
yang searah dengan vector referensi.

# 9. Visualization of Projection

```python
a = np.array([4,2])
b = np.array([1,1])

proj = (np.dot(a,b)/np.dot(b,b))*b

plt.figure(figsize=(6,6))

plt.quiver(0,0,a[0],a[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,b[0],b[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,proj[0],proj[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Vector Projection")

plt.show()
```

Grafik menunjukkan projection vector `a`
ke vector `b`.

# 10. Dimensionality Reduction Concept

Dalam machine learning,
projection digunakan untuk mengurangi dimensi data.

Contoh:
- PCA
- SVD
- Autoencoder

```python
data = np.array([
    [2,3],
    [3,5],
    [4,7],
    [5,9]
])

x = data[:,0]
y = data[:,1]

plt.figure(figsize=(6,5))

plt.scatter(x,y)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("2D Data Representation")

plt.grid()

plt.show()
```

Data dapat direpresentasikan
dalam ruang multidimensi menggunakan vector.
