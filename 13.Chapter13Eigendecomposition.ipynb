```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Introduction to PCA

Principal Component Analysis (PCA)
adalah teknik dimensionality reduction
yang digunakan untuk menyederhanakan data
tanpa kehilangan terlalu banyak informasi.

PCA bekerja dengan mencari:
- arah variasi terbesar data,
- principal components,
- representasi data yang lebih sederhana.

```python
data = np.array([
    [2,3],
    [3,5],
    [4,7],
    [5,9],
    [6,11]
])

print("Dataset:")
print(data)
```

Dataset memiliki dua fitur
yang saling berkorelasi.

# 2. Visualizing the Dataset

```python
x = data[:,0]
y = data[:,1]

plt.figure(figsize=(6,5))

plt.scatter(x,y)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.title("Dataset Visualization")

plt.grid()

plt.show()
```

Scatter plot membantu melihat
hubungan antar fitur.

# 3. Mean Centering

Sebelum PCA dilakukan,
data biasanya di-center
dengan mengurangi mean.

```python
mean = np.mean(data, axis=0)

centered = data - mean

print("Mean:")
print(mean)

print("\nCentered Data:")
print(centered)
```

Mean centering membantu PCA
fokus pada variasi data.

# 4. Covariance Matrix

Covariance matrix digunakan
untuk mengukur hubungan antar fitur.

```python
cov_matrix = np.cov(centered.T)

print("Covariance Matrix:")
print(cov_matrix)
```

Covariance positif menunjukkan
fitur meningkat bersama.

# 5. Heatmap of Covariance Matrix

```python
plt.figure(figsize=(5,4))

plt.imshow(cov_matrix)

plt.colorbar()

plt.title("Covariance Matrix Heatmap")

plt.show()
```

Heatmap membantu visualisasi
hubungan antar fitur.

# 6. Eigenvalues and Eigenvectors

PCA menggunakan eigenvalues
dan eigenvectors
dari covariance matrix.

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

Eigenvector utama menunjukkan
arah variasi terbesar data.

# 7. Principal Components

Principal component pertama
menangkap variasi terbesar data.

```python
pc1 = eigenvectors[:,0]

print("Principal Component 1:")
print(pc1)
```

Principal component digunakan
untuk representasi data baru.

# 8. Visualizing Principal Components

```python
x = centered[:,0]
y = centered[:,1]

pc1 = eigenvectors[:,0]

plt.figure(figsize=(6,6))

plt.scatter(x,y)

plt.quiver(
    0,0,
    pc1[0],pc1[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    width=0.01
)

plt.grid()

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.title("Principal Component")

plt.show()
```

Vector menunjukkan arah
principal component utama.

# 9. Dimensionality Reduction

PCA dapat mengurangi dimensi data
dengan mempertahankan informasi penting.

```python
projected = centered @ pc1

print("Projected Data:")
print(projected)
```

Data dua dimensi
diproyeksikan menjadi satu dimensi.

# 10. Explained Variance

Explained variance menunjukkan
berapa banyak informasi
yang dipertahankan oleh principal component.

```python
variance_ratio = (
    eigenvalues /
    np.sum(eigenvalues)
)

print("Explained Variance Ratio:")
print(variance_ratio)
```

Principal component pertama
biasanya memiliki explained variance terbesar.

# 11. Visualization of Explained Variance

```python
plt.figure(figsize=(6,4))

plt.bar(
    range(len(variance_ratio)),
    variance_ratio
)

plt.xlabel("Principal Component")
plt.ylabel("Explained Variance")

plt.title("Explained Variance Ratio")

plt.show()
```

# 12. PCA in Machine Learning

PCA digunakan dalam:

- dimensionality reduction,
- image compression,
- feature extraction,
- noise reduction,
- data visualization.

```python
X = np.random.randn(100,5)

cov_matrix = np.cov(X.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:")
print(eigenvalues)
```

Eigenvalue membantu menentukan
jumlah principal component
yang penting digunakan.
