```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Introduction to Matrices

Matrix adalah kumpulan angka
yang tersusun dalam bentuk baris dan kolom.

Matrix digunakan untuk:
- menyimpan data,
- transformasi linear,
- machine learning,
- computer graphics,
- deep learning.

```python
A = np.array([
    [1,2,3],
    [4,5,6]
])

print("Matrix A:")
print(A)

print("\nShape:")
print(A.shape)
```

Shape `(2,3)` berarti:

- 2 baris
- 3 kolom

# 2. Matrix Visualization

Matrix dapat divisualisasikan
menggunakan heatmap.

```python
A = np.array([
    [1,2,3],
    [4,5,6]
])

plt.figure(figsize=(5,4))

plt.imshow(A)

plt.colorbar()

plt.title("Matrix Heatmap")

plt.show()
```

Warna pada heatmap menunjukkan
besar nilai pada matrix.

# 3. Matrix Addition

Penjumlahan matrix dilakukan secara element-wise.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

C = A + B

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nA + B:")
print(C)
```

Setiap elemen pada matrix dijumlahkan
dengan elemen yang bersesuaian.

# 4. Scalar Multiplication

Scalar multiplication dilakukan
dengan mengalikan setiap elemen matrix
menggunakan scalar.

```python
A = np.array([
    [1,2],
    [3,4]
])

scalar = 2

result = scalar * A

print("Matrix A:")
print(A)

print("\nResult:")
print(result)
```

Semua elemen matrix dikalikan
dengan scalar 2.

# 5. Matrix Transpose

Transpose dilakukan dengan
menukar baris menjadi kolom.

```python
A = np.array([
    [1,2,3],
    [4,5,6]
])

AT = A.T

print("Matrix A:")
print(A)

print("\nTranspose A:")
print(AT)
```

Transpose matrix mengubah shape:

- dari `(2,3)`
- menjadi `(3,2)`

# 6. Matrix Multiplication

Matrix multiplication berbeda
dengan perkalian element-wise.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

C = A @ B

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nA @ B:")
print(C)
```

Perkalian matrix dilakukan
menggunakan operasi dot product
antara baris dan kolom.

# 7. Identity Matrix

Identity matrix adalah matrix
yang memiliki nilai 1 pada diagonal utama
dan 0 pada elemen lainnya.

```python
I = np.eye(3)

print("Identity Matrix:")
print(I)
```

Identity matrix berfungsi
seperti angka 1 pada perkalian biasa.

# 8. Zero Matrix

Zero matrix adalah matrix
yang seluruh elemennya bernilai 0.

```python
Z = np.zeros((3,3))

print("Zero Matrix:")
print(Z)
```

Zero matrix sering digunakan
untuk inisialisasi data.

# 9. Random Matrix

Random matrix digunakan
untuk simulasi dan machine learning.

```python
R = np.random.randn(4,4)

print("Random Matrix:")
print(R)
```

Nilai matrix dihasilkan secara acak
menggunakan distribusi normal.

# 10. Matrix Multiplication Visualization

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [2,0],
    [1,2]
])

C = A @ B

plt.figure(figsize=(5,4))

plt.imshow(C)

plt.colorbar()

plt.title("Matrix Multiplication Result")

plt.show()
```

Heatmap menunjukkan hasil
perkalian matrix.

# 11. Matrix in Machine Learning

Matrix digunakan hampir di semua bidang machine learning:

- dataset representation,
- neural network,
- image processing,
- PCA,
- recommendation system.

```python
dataset = np.array([
    [170,65],
    [165,60],
    [180,75],
    [175,70]
])

print("Dataset Matrix:")
print(dataset)
```

Setiap baris merepresentasikan data individu,
dan setiap kolom merepresentasikan fitur.
