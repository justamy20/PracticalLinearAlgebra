# 1. Introduction to Matrix Inverse

Inverse matrix digunakan
untuk membalik transformasi linear.

Jika:

A @ A⁻¹ = I

maka A⁻¹ disebut inverse dari matrix A.

```python
A = np.array([
    [2,1],
    [3,4]
])

print("Matrix A:")
print(A)
```

Matrix inverse hanya dapat dihitung
untuk matrix persegi
yang memiliki determinant tidak nol.

# 2. Computing Matrix Inverse

```python
A = np.array([
    [2,1],
    [3,4]
])

A_inv = np.linalg.inv(A)

print("Inverse Matrix:")
print(A_inv)
```

Inverse matrix dihitung
menggunakan fungsi:

`np.linalg.inv()`

# 3. Verifying the Inverse

```python
A = np.array([
    [2,1],
    [3,4]
])

A_inv = np.linalg.inv(A)

result = A @ A_inv

print(result)
```

Hasil perkalian mendekati identity matrix.

# 4. Identity Matrix

Identity matrix adalah matrix
yang memiliki diagonal utama bernilai 1.

```python
I = np.eye(3)

print(I)
```

Identity matrix berfungsi
seperti angka 1
pada perkalian biasa.

# 5. Determinant and Inverse

Determinant menentukan
apakah sebuah matrix memiliki inverse.

```python
A = np.array([
    [2,1],
    [3,4]
])

detA = np.linalg.det(A)

print("Determinant:")
print(detA)
```

Jika determinant ≠ 0,
maka matrix memiliki inverse.

# 6. Singular Matrix

Singular matrix adalah matrix
yang tidak memiliki inverse.

```python
A = np.array([
    [1,2],
    [2,4]
])

detA = np.linalg.det(A)

print("Matrix A:")
print(A)

print("\nDeterminant:")
print(detA)
```

Karena determinant mendekati 0,
matrix termasuk singular matrix.

# 7. Solving Linear Systems Using Inverse

```python
A = np.array([
    [2,1],
    [1,3]
])

b = np.array([
    [8],
    [13]
])

A_inv = np.linalg.inv(A)

x = A_inv @ b

print("Solution:")
print(x)
```

Inverse matrix dapat digunakan
untuk mencari solusi sistem persamaan linear.

# 8. Visualization of Matrix Transformation

```python
v = np.array([1,1])

A = np.array([
    [2,0],
    [0,1]
])

transformed = A @ v

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    transformed[0],transformed[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,3)
plt.ylim(0,3)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Matrix Transformation")

plt.show()
```

Transformasi matrix dapat mengubah
arah dan panjang vector.

# 9. Applying the Inverse Transformation

```python
v = np.array([1,1])

A = np.array([
    [2,0],
    [0,1]
])

A_inv = np.linalg.inv(A)

transformed = A @ v

recovered = A_inv @ transformed

print("Original Vector:")
print(v)

print("\nRecovered Vector:")
print(recovered)
```

Inverse transformation
mengembalikan vector
ke bentuk awal.

# 10. Heatmap Visualization

```python
A = np.array([
    [2,1],
    [3,4]
])

A_inv = np.linalg.inv(A)

plt.figure(figsize=(5,4))

plt.imshow(A_inv)

plt.colorbar()

plt.title("Inverse Matrix Heatmap")

plt.show()
```

Heatmap membantu visualisasi
nilai pada inverse matrix.

# 11. Inverse Matrix in Machine Learning

Inverse matrix digunakan dalam:

- Linear Regression
- Optimization
- Kalman Filter
- Signal Processing
- Computer Graphics

```python
X = np.array([
    [1,1],
    [1,2],
    [1,3]
])

y = np.array([
    [2],
    [4],
    [6]
])

beta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Regression Coefficients:")
print(beta)
```

Perhitungan inverse matrix
digunakan dalam normal equation
pada linear regression.
