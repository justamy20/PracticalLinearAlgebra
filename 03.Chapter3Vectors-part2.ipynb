```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Linear Combination

Linear combination adalah kombinasi beberapa vector
yang dikalikan scalar tertentu.

Bentuk umum:

a*v + b*w

```python
v = np.array([1,2])
w = np.array([3,1])

a = 2
b = -1

linear_combination = a*v + b*w

print("Vector v:")
print(v)

print("\nVector w:")
print(w)

print("\nLinear Combination:")
print(linear_combination)
```

Hasil linear combination diperoleh dari:

2v - w

# 2. Visualization of Linear Combination

```python
v = np.array([1,2])
w = np.array([3,1])

result = 2*v - w

plt.figure(figsize=(6,6))

plt.quiver(0,0,v[0],v[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,w[0],w[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,result[0],result[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(-3,5)
plt.ylim(-3,5)

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Combination")

plt.show()
```

Grafik di atas menunjukkan hasil kombinasi linear
antara dua vector.

# 3. Vector Magnitude

Magnitude atau norm menunjukkan panjang vector.

```python
v = np.array([4,3])

magnitude = np.linalg.norm(v)

print("Vector:")
print(v)

print("\nMagnitude:")
print(magnitude)
```

Magnitude dihitung menggunakan rumus:

sqrt(x² + y²)

# 4. Unit Vector

Unit vector adalah vector
yang memiliki panjang 1.

```python
v = np.array([4,3])

unit_vector = v / np.linalg.norm(v)

print("Vector:")
print(v)

print("\nUnit Vector:")
print(unit_vector)

print("\nNorm Unit Vector:")
print(np.linalg.norm(unit_vector))
```

Normalisasi vector dilakukan dengan membagi
vector menggunakan magnitude-nya.

# 5. Dot Product Review

Dot product digunakan untuk mengukur
hubungan antar vector.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

dot = np.dot(a,b)

print("Dot Product:")
print(dot)
```

Jika dua vector memiliki arah mirip,
maka dot product bernilai positif besar.

# 6. Cosine Similarity

Cosine similarity digunakan untuk mengukur
kemiripan arah antar vector.

```python
v1 = np.array([1,2,3])
v2 = np.array([2,4,6])

cosine_similarity = np.dot(v1,v2) / (
    np.linalg.norm(v1) * np.linalg.norm(v2)
)

print("Cosine Similarity:")
print(cosine_similarity)
```

Nilai cosine similarity mendekati 1
menunjukkan bahwa kedua vector
memiliki arah yang sangat mirip.

# 7. Orthogonal Vectors

Dua vector orthogonal memiliki dot product nol.

```python
a = np.array([1,0])
b = np.array([0,1])

dot = np.dot(a,b)

print("Dot Product:")
print(dot)
```

Karena hasil dot product bernilai 0,
maka kedua vector orthogonal.

# 8. Projection of One Vector onto Another

Projection digunakan untuk mengetahui
komponen sebuah vector terhadap vector lain.

```python
a = np.array([4,2])
b = np.array([1,1])

projection = (
    np.dot(a,b) / np.dot(b,b)
) * b

print("Vector a:")
print(a)

print("\nVector b:")
print(b)

print("\nProjection of a onto b:")
print(projection)
```

Projection menghasilkan bagian vector `a`
yang searah dengan vector `b`.

# 9. Visualization of Projection

```python
a = np.array([4,2])
b = np.array([1,1])

proj = (np.dot(a,b)/np.dot(b,b))*b

plt.figure(figsize=(6,6))

plt.quiver(0,0,a[0],a[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,b[0],b[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,proj[0],proj[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Vector Projection")

plt.show()
```

Grafik menunjukkan projection vector `a`
terhadap vector `b`.

# 10. Linear Independence

Sekumpulan vector dikatakan independent
jika tidak dapat dibentuk
dari kombinasi linear vector lain.

```python
v1 = np.array([1,0])
v2 = np.array([0,1])

matrix = np.column_stack((v1,v2))

rank = np.linalg.matrix_rank(matrix)

print("Matrix:")
print(matrix)

print("\nRank:")
print(rank)
```

Karena rank matrix = jumlah vector,
maka kedua vector independent.

# 11. Span of Vectors

Span adalah seluruh kemungkinan vector
yang dapat dibentuk
dari kombinasi linear beberapa vector.

```python
v1 = np.array([1,0])
v2 = np.array([0,1])

a = 2
b = 3

result = a*v1 + b*v2

print("Result:")
print(result)
```

Vector `[2,3]` berada pada span
dari vector `[1,0]` dan `[0,1]`.
