```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Introduction to Linear Least Squares

Linear least squares digunakan
untuk mencari garis terbaik
yang meminimalkan error
antara data dan prediksi model.

Tujuan least squares adalah
meminimalkan jumlah kuadrat error.

# 2. Creating Sample Data

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

print("x:")
print(x)

print("\ny:")
print(y)
```

Dataset memiliki hubungan linear sederhana.

# 3. Visualizing the Dataset

```python
plt.figure(figsize=(6,5))

plt.scatter(x,y)

plt.xlabel("x")
plt.ylabel("y")

plt.title("Dataset Visualization")

plt.grid()

plt.show()
```

Scatter plot membantu melihat
pola hubungan antar data.

# 4. Linear Regression Model

Model regresi linear:

y = mx + b

```python
# 4. Linear Regression Model

Model regresi linear:

y = mx + b
```

m → slope  
b → intercept

m, b = np.polyfit(x,y,1)

print("Slope:")
print(m)

print("\nIntercept:")
print(b)

NumPy menyediakan fungsi
untuk menghitung garis regresi linear.

# 5. Predicting Values

```python
y_pred = m*x + b

print("Predicted Values:")
print(y_pred)
```

Prediksi diperoleh
menggunakan model regresi linear.

# 6. Visualizing the Regression Line

```python
plt.figure(figsize=(6,5))

plt.scatter(x,y)

plt.plot(x,y_pred)

plt.xlabel("x")
plt.ylabel("y")

plt.title("Linear Regression")

plt.grid()

plt.show()
```

Garis regresi menunjukkan
hubungan linear terbaik
antara x dan y.

# 7. Residual Error

Residual error adalah
selisih antara data asli
dan hasil prediksi.

```python
residuals = y - y_pred

print("Residuals:")
print(residuals)
```

Residual kecil menunjukkan
model memiliki prediksi yang baik.

# 8. Sum of Squared Errors (SSE)

SSE digunakan untuk mengukur
total error model.

```python
sse = np.sum(residuals**2)

print("Sum of Squared Errors:")
print(sse)
```

Semakin kecil SSE,
semakin baik model regresi.

# 9. Least Squares Solution

Solusi least squares dapat ditulis sebagai:

β = (XᵀX)⁻¹Xᵀy

```python
X = np.column_stack((x,np.ones(len(x))))

beta = (
    np.linalg.inv(X.T @ X)
    @ X.T
    @ y
)

print("Beta:")
print(beta)
```

Matrix equation digunakan
untuk menghitung parameter regresi.

# 10. Visualization of Residuals

```python
plt.figure(figsize=(6,5))

plt.scatter(x,y)

plt.plot(x,y_pred)

for i in range(len(x)):
    plt.plot(
        [x[i],x[i]],
        [y[i],y_pred[i]]
    )

plt.xlabel("x")
plt.ylabel("y")

plt.title("Residual Errors")

plt.grid()

plt.show()
```

Garis vertikal menunjukkan
besar residual error.

# 11. Applications in Machine Learning
