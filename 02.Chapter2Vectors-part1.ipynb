```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Creating and Visualizing Vectors in NumPy

Dalam linear algebra, vector adalah kumpulan angka
yang tersusun secara teratur.

Dua sifat utama vector:

- Dimensionality → jumlah elemen vector
- Orientation → row vector atau column vector

Secara umum, vector direpresentasikan
menggunakan NumPy array.

```python
v = np.array([2, 4])

print("Vector v:")
print(v)

print("\nShape:")
print(v.shape)
```

Vector di atas memiliki:

- 2 elemen
- shape `(2,)`

yang berarti vector satu dimensi.

# 2. Visualizing a Vector

Vector dapat divisualisasikan sebagai panah
dari titik asal `(0,0)` menuju titik tertentu.

```python
v = np.array([2,4])

plt.figure(figsize=(5,5))

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Visualization of Vector v")

plt.show()
```

Grafik di atas menunjukkan vector `v`
bergerak dari titik `(0,0)` menuju `(2,4)`.

# 3. Row Vector dan Column Vector

Vector dapat direpresentasikan sebagai:

- Row Vector
- Column Vector

```python
row_vector = np.array([[1,2,3]])
column_vector = np.array([[1],
                          [2],
                          [3]])

print("Row Vector:")
print(row_vector)

print("\nShape:")
print(row_vector.shape)

print("\nColumn Vector:")
print(column_vector)

print("\nShape:")
print(column_vector.shape)
```

Shape row vector adalah `(1,3)`:

- 1 baris
- 3 kolom

Shape column vector adalah `(3,1)`:

- 3 baris
- 1 kolom

# 4. Vector Addition

Penjumlahan vector dilakukan secara element-wise.

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

hasil = v1 + v2

print("Vector 1:")
print(v1)

print("\nVector 2:")
print(v2)

print("\nHasil Penjumlahan:")
print(hasil)
```

Perhitungan:

- 1 + 4 = 5
- 2 + 5 = 7
- 3 + 6 = 9

# 5. Visualisasi Penjumlahan Vector

```python
v1 = np.array([1,2])
v2 = np.array([2,1])

v3 = v1 + v2

plt.figure(figsize=(6,6))

plt.quiver(0,0,v1[0],v1[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,v2[0],v2[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,v3[0],v3[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()
plt.title("Vector Addition")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

Vector hasil penjumlahan merupakan gabungan
dari kedua vector sebelumnya.

# 6. Scalar Multiplication

Scalar multiplication dilakukan dengan mengalikan
setiap elemen vector menggunakan scalar.

```python
v = np.array([1,2])

scalar = 3

hasil = scalar * v

print("Vector:")
print(v)

print("\nScalar:")
print(scalar)

print("\nHasil:")
print(hasil)
```

Scalar multiplication mengubah panjang vector
tetapi tidak mengubah arah vector
(jika scalar bernilai positif).

# 7. Visualisasi Scalar Multiplication

```python
v = np.array([1,2])

scaled_v = 3*v

plt.figure(figsize=(6,6))

plt.quiver(0,0,v[0],v[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,scaled_v[0],scaled_v[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(0,5)
plt.ylim(0,7)

plt.grid()
plt.title("Scalar Multiplication")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

# 8. Dot Product

Dot product merupakan operasi penting
dalam linear algebra.

Dot product dihitung dengan:

- mengalikan elemen yang bersesuaian,
- menjumlahkan hasil perkalian.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

dot_product = np.dot(a,b)

print("Vector a:")
print(a)

print("\nVector b:")
print(b)

print("\nDot Product:")
print(dot_product)
```

Perhitungan:

(1×4) + (2×5) + (3×6)

= 4 + 10 + 18

= 32

# 9. Geometric Interpretation of Dot Product

Dot product dapat digunakan untuk mengukur
hubungan antar vector.

- Nilai positif → arah vector mirip
- Nilai nol → orthogonal
- Nilai negatif → arah berlawanan

```python
v1 = np.array([2,1])
v2 = np.array([1,2])

plt.figure(figsize=(6,6))

plt.quiver(0,0,v1[0],v1[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,v2[0],v2[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(0,4)
plt.ylim(0,4)

plt.grid()
plt.title("Dot Product Geometry")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

# 10. Vector Norm

Norm digunakan untuk menghitung panjang vector.

```python
v = np.array([3,4])

norm = np.linalg.norm(v)

print("Vector:")
print(v)

print("\nNorm:")
print(norm)
```

Perhitungan norm:

sqrt(3² + 4²)

= sqrt(9 + 16)

= sqrt(25)

= 5

# 11. Orthogonal Vectors

Dua vector dikatakan orthogonal
jika dot product bernilai nol.

```python
a = np.array([1,0])
b = np.array([0,1])

dot = np.dot(a,b)

print("Dot Product:")
print(dot)
```

Karena hasil dot product adalah 0,
maka kedua vector orthogonal.
