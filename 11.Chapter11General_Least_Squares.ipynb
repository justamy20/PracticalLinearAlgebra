```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Introduction to Eigenvalues and Eigenvectors

Eigenvector adalah vector
yang arahannya tetap
setelah transformasi linear dilakukan.

Eigenvalue menunjukkan
besar scaling pada eigenvector.

```python
A = np.array([
    [2,1],
    [1,2]
])

print("Matrix A:")
print(A)
```

Eigenvector hanya berubah panjangnya,
tetapi arah vector tetap sama.

# 2. Computing Eigenvalues and Eigenvectors

```python
A = np.array([
    [2,1],
    [1,2]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

Setiap eigenvalue
berpasangan dengan eigenvector tertentu.

# 3. Verifying the Eigenvector Equation

Persamaan eigenvector:

A v = λ v

```python
A = np.array([
    [2,1],
    [1,2]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

v = eigenvectors[:,0]
lam = eigenvalues[0]

left_side = A @ v
right_side = lam * v

print("A @ v:")
print(left_side)

print("\nlambda * v:")
print(right_side)
```

Hasil kedua sisi persamaan
memiliki nilai yang sama.

# 4. Visualizing Eigenvectors

```python
A = np.array([
    [2,1],
    [1,2]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

v1 = eigenvectors[:,0]
v2 = eigenvectors[:,1]

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    v1[0],v1[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    v2[0],v2[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(-1,1)
plt.ylim(-1,1)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Eigenvectors")

plt.show()
```

# 5. Matrix Transformation Using Eigenvectors

```python
A = np.array([
    [2,1],
    [1,2]
])

v = np.array([1,1])

transformed = A @ v

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    transformed[0],transformed[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Eigenvector Transformation")

plt.show()
```

Vector berubah panjang
tetapi tetap berada
pada arah yang sama.

# 6. Diagonal Matrix

Diagonal matrix memiliki:
- elemen diagonal utama,
- elemen lain bernilai 0.

```python
D = np.diag([3,1])

print(D)
```

Diagonal matrix sering digunakan
dalam eigendecomposition.

# 7. Eigendecomposition

Eigendecomposition ditulis sebagai:

A = VDV⁻¹

```python
A = np.array([
    [2,1],
    [1,2]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

D = np.diag(eigenvalues)
V = eigenvectors
V_inv = np.linalg.inv(V)

reconstructed = V @ D @ V_inv

print("Reconstructed Matrix:")
print(reconstructed)
```

Matrix hasil reconstruction
mendekati matrix asli.

# 8. Symmetric Matrix and Eigenvectors

Matrix symmetric memiliki:
- eigenvalue real,
- eigenvector orthogonal.

```python
A = np.array([
    [2,1],
    [1,2]
])

print("Transpose:")
print(A.T)
```

Karena A = Aᵀ,
matrix termasuk symmetric matrix.

# 10. PCA and Eigenvectors

Principal Component Analysis (PCA)
menggunakan eigenvector
untuk menentukan arah variasi terbesar data.

```python
data = np.array([
    [2,3],
    [3,5],
    [4,7],
    [5,9]
])

cov_matrix = np.cov(data.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```

Eigenvector utama digunakan
sebagai principal component.
