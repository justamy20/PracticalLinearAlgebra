# 1. Introduction to Optimization

Optimization digunakan
untuk mencari parameter terbaik
yang meminimalkan error model.

Optimization merupakan inti
dari machine learning modern.

Tujuan optimization adalah:
- meminimalkan cost function,
- meningkatkan akurasi model,
- mencari solusi optimal.

# 2. Cost Function

Cost function mengukur
besar error model.

Contoh fungsi kuadrat:

f(x) = x²

```python
x = np.linspace(-5,5,100)

y = x**2

plt.figure(figsize=(6,5))

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("Cost Function")

plt.grid()

plt.show()
```

Nilai minimum fungsi berada
di titik x = 0.

# 3. Gradient Descent

Gradient descent digunakan
untuk mencari minimum function
secara iteratif.

Rumus gradient descent:

x = x - α∇f(x)

α → learning rate  
∇f(x) → gradient

```python
x = 4

learning_rate = 0.1

for i in range(10):

    gradient = 2*x

    x = x - learning_rate * gradient

    print(f"Iteration {i+1}: x = {x}")
```

Nilai x bergerak mendekati
titik minimum function.

# 4. Visualizing Gradient Descent

```python
x_values = np.linspace(-5,5,100)

y_values = x_values**2

x = 4

positions = []

learning_rate = 0.1

for i in range(10):

    positions.append(x)

    gradient = 2*x

    x = x - learning_rate * gradient

plt.figure(figsize=(6,5))

plt.plot(x_values,y_values)

plt.scatter(
    positions,
    np.array(positions)**2
)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("Gradient Descent Optimization")

plt.grid()

plt.show()
```

Titik-titik menunjukkan
langkah optimization
menuju minimum function.

# 5. Learning Rate

Learning rate menentukan
besar langkah update parameter.

```python
learning_rates = [0.01,0.1,0.5]

for lr in learning_rates:

    x = 4

    for i in range(10):

        gradient = 2*x

        x = x - lr * gradient

    print(f"Learning Rate {lr}: x = {x}")
```

Learning rate terlalu kecil
membuat proses lambat.

Learning rate terlalu besar
dapat menyebabkan instability.

# 6. Optimization in Linear Regression

Optimization digunakan
untuk mencari parameter regresi terbaik.

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

m = 0
b = 0

learning_rate = 0.01

for i in range(100):

    y_pred = m*x + b

    dm = (-2/len(x)) * np.sum(x*(y-y_pred))
    db = (-2/len(x)) * np.sum(y-y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db

print("Slope:")
print(m)

print("\nIntercept:")
print(b)
```

Gradient descent membantu
mencari parameter optimal
secara otomatis.

# 7. Loss Visualization

```python
losses = []

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

m = 0
b = 0

learning_rate = 0.01

for i in range(100):

    y_pred = m*x + b

    loss = np.mean((y-y_pred)**2)

    losses.append(loss)

    dm = (-2/len(x)) * np.sum(x*(y-y_pred))
    db = (-2/len(x)) * np.sum(y-y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db

plt.figure(figsize=(6,5))

plt.plot(losses)

plt.xlabel("Iteration")
plt.ylabel("Loss")

plt.title("Loss During Optimization")

plt.grid()

plt.show()
```

Loss menurun seiring
proses optimization berjalan.

# 8. Convex Function

Convex function memiliki
satu titik minimum global.

```python
x = np.linspace(-5,5,100)

y = x**2 + 2

plt.figure(figsize=(6,5))

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("Convex Function")

plt.grid()

plt.show()
```

Convex optimization lebih mudah
karena memiliki minimum global tunggal.

# 9. Non-Convex Function

```python
x = np.linspace(-5,5,100)

y = np.sin(x) + 0.1*x**2

plt.figure(figsize=(6,5))

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("Non-Convex Function")

plt.grid()

plt.show()
```

Non-convex function memiliki
banyak local minimum.

# 10. Applications in Machine Learning

Optimization digunakan dalam:

```python
weights = np.random.randn(5)

gradient = np.random.randn(5)

learning_rate = 0.01

updated_weights = (
    weights -
    learning_rate * gradient
)

print("Updated Weights:")
print(updated_weights)
```
