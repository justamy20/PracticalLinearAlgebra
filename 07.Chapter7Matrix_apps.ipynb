```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Matrices as Data Representation

Dalam data science,
dataset biasanya direpresentasikan
dalam bentuk matrix.

- Baris → observasi/data
- Kolom → fitur/feature

```python
dataset = np.array([
    [170,65],
    [165,60],
    [180,75],
    [175,70]
])

print("Dataset Matrix:")
print(dataset)

print("\nShape:")
print(dataset.shape)
```

Dataset memiliki:
- 4 observasi
- 2 fitur

# 2. Visualizing Dataset Matrix

```python
height = dataset[:,0]
weight = dataset[:,1]

plt.figure(figsize=(6,5))

plt.scatter(height,weight)

plt.xlabel("Height")
plt.ylabel("Weight")

plt.title("Dataset Visualization")

plt.grid()

plt.show()
```

Scatter plot membantu memahami
hubungan antar fitur pada dataset.

# 3. Matrix Transformation

Matrix transformation digunakan untuk:
- scaling,
- rotation,
- reflection,
- shearing.

```python
vector = np.array([1,1])

A = np.array([
    [2,0],
    [0,1]
])

transformed = A @ vector

print("Original Vector:")
print(vector)

print("\nTransformed Vector:")
print(transformed)
```

Transformation matrix mengubah
ukuran dan arah vector.

# 4. Visualizing Matrix Transformation

```python
vector = np.array([1,1])

A = np.array([
    [2,0],
    [0,1]
])

transformed = A @ vector

plt.figure(figsize=(6,6))

plt.quiver(0,0,
           vector[0],vector[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,
           transformed[0],transformed[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(0,3)
plt.ylim(0,3)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Matrix Transformation")

plt.show()
```

Grafik menunjukkan vector
sebelum dan sesudah transformasi matrix.

# 5. Rotation Matrix

Rotation matrix digunakan
untuk memutar vector.

```python
theta = np.pi / 4

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

v = np.array([1,0])

rotated = R @ v

print("Rotated Vector:")
print(rotated)
```

Rotation matrix memutar vector
sebesar sudut tertentu.

# 6. Visualization of Rotation

```python
theta = np.pi / 4

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

v = np.array([1,0])

rotated = R @ v

plt.figure(figsize=(6,6))

plt.quiver(0,0,v[0],v[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.quiver(0,0,
           rotated[0],
           rotated[1],
           angles='xy',
           scale_units='xy',
           scale=1)

plt.xlim(-1,2)
plt.ylim(-1,2)

plt.grid()

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Rotation Matrix")

plt.show()
```

Grafik menunjukkan vector
yang diputar 45 derajat.

# 7. Covariance Matrix

Covariance matrix digunakan
untuk mengukur hubungan antar fitur.

```python
data = np.array([
    [170,65],
    [165,60],
    [180,75],
    [175,70]
])

cov_matrix = np.cov(data.T)

print("Covariance Matrix:")
print(cov_matrix)
```

Covariance positif menunjukkan
dua fitur meningkat bersama.

# 8. Heatmap of Covariance Matrix

```python
plt.figure(figsize=(5,4))

plt.imshow(cov_matrix)

plt.colorbar()

plt.title("Covariance Matrix Heatmap")

plt.show()
```

Heatmap membantu visualisasi
hubungan antar fitur.

# 9. Image Representation Using Matrix

Dalam computer vision,
gambar direpresentasikan sebagai matrix pixel.

```python
image_matrix = np.random.rand(10,10)

plt.figure(figsize=(5,5))

plt.imshow(image_matrix, cmap='gray')

plt.title("Image Matrix Representation")

plt.show()
```

Setiap elemen matrix
merepresentasikan intensitas pixel.

# 10. Matrix in Neural Networks

Neural network menggunakan matrix
untuk:
- input data,
- weight,
- transformasi layer.

```python
inputs = np.array([
    [1,2],
    [3,4]
])

weights = np.array([
    [0.5,1.0],
    [1.5,2.0]
])

output = inputs @ weights

print("Neural Network Output:")
print(output)
```

Perkalian matrix digunakan
untuk menghitung output neural network.

# 11. Linear Regression Example

Linear regression menggunakan matrix
untuk menghitung parameter model.

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

plt.figure(figsize=(6,5))

plt.plot(x,y,'o-')

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Linear Regression Example")

plt.grid()

plt.show()
```

Grafik menunjukkan hubungan linear
antara x dan y.
