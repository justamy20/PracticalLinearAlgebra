```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Introduction to Row Reduction

Row reduction digunakan
untuk menyederhanakan matrix
menggunakan operasi baris elementer.

Tujuan utama:
- mencari solusi sistem linear,
- menghitung rank,
- menentukan inverse matrix.

```python
A = np.array([
    [1,2],
    [3,4]
])

print("Matrix A:")
print(A)
```

Operasi baris elementer meliputi:
- menukar baris,
- mengalikan baris,
- menjumlahkan baris.

# 2. Gaussian Elimination

Gaussian elimination digunakan
untuk mengubah matrix
menjadi echelon form.

```python
A = np.array([
    [1,2],
    [3,4]
], dtype=float)

A[1] = A[1] - 3*A[0]

print("Row Reduced Matrix:")
print(A)
```

Baris kedua dikurangi
3 kali baris pertama.

# 3. Echelon Form

Echelon form memiliki:
- leading coefficient,
- elemen di bawah leading coefficient bernilai 0.

```python
A = np.array([
    [1,2,3],
    [0,1,4],
    [0,0,1]
])

print(A)
```

Matrix di atas sudah berbentuk echelon form.

# 4. Reduced Row Echelon Form (RREF)

RREF merupakan bentuk lanjutan
dari echelon form.

A = np.array([
    [1,2],
    [3,4]
], dtype=float)

A[1] = A[1] - 3*A[0]
A[1] = A[1] / A[1,1]
A[0] = A[0] - 2*A[1]

print(A)

Matrix hasil reduction
lebih mudah digunakan
untuk mencari solusi sistem linear.

# 5. Matrix Rank

Rank menunjukkan jumlah
baris atau kolom independent.

```python
A = np.array([
    [1,2],
    [2,4]
])

rank = np.linalg.matrix_rank(A)

print("Matrix A:")
print(A)

print("\nRank:")
print(rank)
```

Karena rank = 1,
matrix memiliki informasi redundant.

# 6. Full Rank Matrix

Matrix full rank memiliki:
- seluruh baris independent,
- seluruh kolom independent.

```python
A = np.array([
    [1,2],
    [3,4]
])

rank = np.linalg.matrix_rank(A)

print("Rank:")
print(rank)
```

Karena rank = jumlah dimensi matrix,
maka matrix termasuk full rank.

# 7. Solving Linear Systems

Matrix digunakan
untuk menyelesaikan
sistem persamaan linear.

```python
A = np.array([
    [2,1],
    [1,3]
])

b = np.array([8,13])

x = np.linalg.solve(A,b)

print("Solution:")
print(x)
```

Hasil menunjukkan solusi
dari sistem persamaan linear.

# 8. Visualization of Linear Equations

```python
x = np.linspace(0,10,100)

y1 = 8 - 2*x
y2 = (13 - x)/3

plt.figure(figsize=(6,5))

plt.plot(x,y1,label='2x + y = 8')
plt.plot(x,y2,label='x + 3y = 13')

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()

plt.xlabel("x")
plt.ylabel("y")

plt.legend()

plt.title("System of Linear Equations")

plt.show()
```

Titik perpotongan kedua garis
merupakan solusi sistem linear.

# 9. Inconsistent System

Sistem linear inconsistent
tidak memiliki solusi.

```python
x = np.linspace(0,10,100)

y1 = 5 - x
y2 = 7 - x

plt.figure(figsize=(6,5))

plt.plot(x,y1,label='x + y = 5')
plt.plot(x,y2,label='x + y = 7')

plt.grid()

plt.xlabel("x")
plt.ylabel("y")

plt.legend()

plt.title("Inconsistent System")

plt.show()
```

Dua garis paralel
tidak memiliki titik potong,
sehingga sistem tidak memiliki solusi.

# 10. Matrix Rank Visualization

Heatmap membantu visualisasi
struktur matrix.

# 11. Applications in Machine Learning

Row reduction dan matrix rank
digunakan dalam:

- linear regression,
- PCA,
- optimization,
- neural networks,
- data compression.

```python
X = np.array([
    [1,1],
    [1,2],
    [1,3]
])

rank = np.linalg.matrix_rank(X)

print("Matrix Rank:")
print(rank)
```

Rank matrix membantu menentukan
apakah fitur dalam dataset
bersifat redundant atau tidak.
