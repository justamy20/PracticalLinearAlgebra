# 1. Introduction to Singular Value Decomposition

Singular Value Decomposition (SVD)
adalah metode dekomposisi matrix menjadi:

A = UΣVᵀ

Komponen SVD:
- U → left singular vectors
- Σ → singular values
- Vᵀ → right singular vectors

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

print("Matrix A:")
print(A)
```

SVD dapat diterapkan
pada matrix persegi
maupun non-persegi.

# 2. Computing SVD

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

U, S, VT = np.linalg.svd(A)

print("Matrix U:")
print(U)

print("\nSingular Values:")
print(S)

print("\nMatrix VT:")
print(VT)
```

NumPy menyediakan fungsi:

`np.linalg.svd()`

# 3. Singular Values

Singular values menunjukkan
besar kontribusi masing-masing dimensi.

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

U, S, VT = np.linalg.svd(A)

print("Singular Values:")
print(S)
```

Singular values selalu bernilai positif.

# 4. Creating Sigma Matrix

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

U, S, VT = np.linalg.svd(A)

Sigma = np.zeros((3,2))

Sigma[:2,:2] = np.diag(S)

print(Sigma)
```

Matrix Σ berisi singular values
pada diagonal utama.

# 5. Reconstructing the Matrix

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

U, S, VT = np.linalg.svd(A)

Sigma = np.zeros((3,2))
Sigma[:2,:2] = np.diag(S)

A_reconstructed = U @ Sigma @ VT

print("Reconstructed Matrix:")
print(A_reconstructed)
```

Matrix hasil reconstruction
mendekati matrix asli.

# 6. Visualization of Singular Values

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

U, S, VT = np.linalg.svd(A)

plt.figure(figsize=(6,4))

plt.plot(S,'o-')

plt.xlabel("Index")
plt.ylabel("Singular Value")

plt.title("Singular Values")

plt.grid()

plt.show()
```

Grafik menunjukkan besar
setiap singular value.

# 7. Dimensionality Reduction

SVD dapat digunakan
untuk mengurangi dimensi data.

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

U, S, VT = np.linalg.svd(A)

k = 1

A_approx = (
    U[:,:k] @
    np.diag(S[:k]) @
    VT[:k,:]
)

print("Low Rank Approximation:")
print(A_approx)
```

Low rank approximation
menyimpan informasi utama data
dengan dimensi lebih kecil.

# 9. Image Compression Using SVD

SVD digunakan
untuk kompresi gambar
dengan mempertahankan informasi penting.

```python
image = np.random.rand(50,50)

U, S, VT = np.linalg.svd(image)

k = 5

compressed = (
    U[:,:k] @
    np.diag(S[:k]) @
    VT[:k,:]
)

plt.figure(figsize=(5,5))

plt.imshow(compressed, cmap='gray')

plt.title("Compressed Image")

plt.show()
```

SVD dapat mengurangi ukuran data
tanpa menghilangkan terlalu banyak informasi.

# 11. Rank Approximation

Rank approximation digunakan
untuk menyederhanakan matrix
dengan mempertahankan informasi penting.

```python
A = np.random.randn(10,10)

U, S, VT = np.linalg.svd(A)

k = 2

A2 = (
    U[:,:k] @
    np.diag(S[:k]) @
    VT[:k,:]
)

print("Original Shape:")
print(A.shape)

print("\nApproximation Shape:")
print(A2.shape)
```
